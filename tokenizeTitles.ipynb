```python
import pandas as pd
from tqdm import tqdm
tqdm.pandas()
import json
```

```python
df = pd.read_csv("NPOPlayer.csv", sep=";")
```

```python
mediaID_MainTitleTokens = {}

for index, row in tqdm(df.iterrows()):
    mediaID_MainTitleTokens[row["mediaID"]] = str(row["mainTitle"]).lower().split(" ")

mediaID_MainTitleTokens
```

```python
def save_activities(dictname):
    with open(f'{dictname}.json', 'w') as outfile:
        json.dump(dictname, outfile)
save_activities(mediaID_MainTitleTokens)
```
